Importing required libraries

```
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```
from sklearn.datasets import load_iris
```

Read the of the data into a dataframe

```
iris = load_iris()
```

```
iris_pd = pd.DataFrame(iris.data,columns=iris.feature_names)
iris_pd
```

```
iris.target_names
```

```
iris_pd['targets'] = iris.target
```

```
iris_pd
```

```
iris_pd = iris_pd.sample(frac=1)
iris_pd
```

Making of feature matrix and target matrix

```
X = iris_pd.loc[:, iris_pd.columns != 'targets'].to_numpy()
y = iris_pd['targets'].to_numpy()
```

```
y
```

To make all features in same range we import StandardScalar of sklearn to do this.



```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

Importing SVM, Random forest, Logistic Regression classifiers for model selection



```
from sklearn import svm
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

Running Grid search for finding optimal model for prediction

```
grid = {
    'svm':{
        'model': svm.SVC(gamma = 'auto'),
        'par':{
            'C':[1,10,100],
            'kernel':['rbf','linear','poly']
        }
    },
    'RandomForest':{
        'model': RandomForestClassifier(),
        'par':{
            'n_estimators':[20,40,60,80,100]
        }
    },
    'Logistic':{
        'model': LogisticRegression(),
        'par':{
            'C':[5,10,25,50,75,100]
        }    
    }
}
```

```
scores = []
for model_name,mp in grid.items():
  model = GridSearchCV(mp['model'],mp['par'],cv= 5)
  model.fit(X,y)
  scores.append({
      'model':model_name,
      'best_score': model.best_score_,
      'best_param': model.best_params_,
  })
```

```
k = pd.DataFrame(scores,columns=['model','best_score','best_param'])
k
```

As we can see we get highest accuracy when model is SVM with accuracy of 0.973333
